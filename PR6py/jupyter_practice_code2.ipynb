```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

valorant = pd.read_csv("valorant-stats_PALOMALO_SILNO.csv")
```

```python
guns_prices = valorant["Price"]

guns_damage = valorant["HDMG_0"]

plt.plot(guns_prices, guns_damage, color="pink", marker="o", markersize=8, label="График соотношения цены и урона в голову")
plt.title("График соотношения цены и урона в голову")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```

Создаём обычный линейный график, с соотношением цены оружия и урона в голову, видим, что урон +- растёт с увеличением цены

```python
guns_damage = valorant["BDMG_0"]

plt.plot(guns_prices, guns_damage, color="blue", marker="o", markersize=8, label="График соотношения цены и урона в тело")
plt.title("График соотношения цены и урона в тело")
plt.legend()
plt.xlabel("УРОН")
plt.ylabel("ЦЕНА")
plt.show()
```

Создаём ещё один обычный линейный график, с соотношением цены оружия и урона в тело, видим, что урон +- растёт с увеличением цены

```python
guns_damage = valorant["LDMG_0"]

plt.plot(guns_prices, guns_damage, color="red", marker="o", markersize=8, label="График соотношения цены и урона в ноги")
plt.title("График соотношения цены и урона в ноги")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```

Создаём ещё один обычный линейный график, с соотношением цены оружия и урона в ноги, видим, что урон +- растёт с увеличением цены

Да, у меня получилось три, почти одинаковых графика, но по факту они должны быть разными, и показывают относительно разные зависимости. Я не придумаю ничего лучше, чем это

```python
guns_damage_in_body = valorant["Wall Penetration"]
guns_names = valorant["Name"]

plt.figure(figsize=(15, 5))
plt.bar(guns_names, guns_damage_in_body, color="purple", label="График простреливаемости стен")
plt.title("График простреливаемости стен")
plt.legend()
plt.xlabel("Имя оружия")
plt.ylabel("Простреливаемость стен")
plt.show()
```

Создаём столбчатую диаграмму, видим что самая большая сила прострела у operator, ares и sheriff 

```python
guns_damage_in_body = valorant["Fire Rate"]
guns_names = valorant["Name"]

plt.figure(figsize=(15, 5))
plt.bar(guns_names, guns_damage_in_body, color="pink", label="График скорости стрельбы")
plt.title("График скорости стрельбы")
plt.legend()
plt.xlabel("Имя оружия")
plt.ylabel("Скорость стрельбы")
plt.show()
```

Создаём ещё одну столбчатую диаграмму, видим что самая большая скорость стрельбы у stinger, очень круто для ПП 

```python
guns_damage_in_body = valorant["BDMG_0"]
guns_names = valorant["Name"]

plt.figure(figsize=(15, 6))
plt.bar(guns_names, guns_damage_in_body, color="b", label="График урона в тело")
plt.title("График урона в тело")
plt.legend()
plt.xlabel("Имя оружия")
plt.ylabel("урон в тело")
plt.show()
```

Создаём ещё одну столбчатую диаграмму, видим что самый большой урон в тело у Operator, что довольно логично, потому что это снайперская винтовка 

```python
rifles_count = (valorant["Weapon Type"] == "Rifle").sum()
sidearm_count = (valorant["Weapon Type"] == "Sidearm").sum()
x = (rifles_count, sidearm_count)
plt.figure(figsize=(10,10))
plt.pie(x, labels=("Винтовки", "Пистолеты"), autopct="%1.1f%%")
plt.legend()
plt.title("Диаграмма соотношения количества винтовок и пистолетов")
plt.show()
```

Создаём ещё одну круговую диаграмму с соотношением количества пистолетов и винтовок в игре, видим что пистолетов больше, это мега круто, классная диаграмма, крутая статистика

```python
rifles_count = (valorant["Weapon Type"] == "SMG").sum()
sidearm_count = (valorant["Weapon Type"] == "Shotgun").sum()
x = (rifles_count, sidearm_count)
plt.figure(figsize=(10,10))
plt.pie(x, labels=("Пистолеты пулемёты", "Дробовики"), autopct="%1.1f%%")
plt.legend()
plt.title("Диаграмма соотношения количества пистолетов пулемётов и дробавиков")
plt.show()
```

Создаём ещё одну круговую диаграмму с соотношением количества дробовиков и ПП в игре, видим что их количество одинаково, круто

```python
plt.scatter(valorant["Fire Rate"], valorant["Wall Penetration"])
plt.title("График связи скорости стрельбы и простреливаемости стен")
plt.xlabel("Скорость стрельбы")
plt.ylabel("Простреливаемость стен")
plt.show()
```

Получаем график рассеивания, видим, что в нём отсутствует связь, значит здесь что рост скорострельности, никак не влияет на рост простреливаемости, и наоборот

```python
plt.scatter(valorant["Fire Rate"], valorant["Magazine Capacity"])
plt.title("График связи скорости стрельбы и объёма магазина")
plt.xlabel("Скорость стрельбы")
plt.ylabel("Объём магазина")
plt.show()
```

Получаем ещё один график рассеивания, видим, что в нём есть слабая связь при росте скорострельности, происходит рост объёма магазина. Что в целом довольно логично, но конкретно здесь связь слабая

```python
plt.figure(figsize=(10,4))
plt.hist(valorant["Fire Rate"], bins = 18, edgecolor = "black", color = "yellow")
plt.title("Распределение частоты огня")
plt.xlabel("Частота огня")
plt.ylabel("Количество значения")
plt.show()
```

Создаём гистограмму в которой будем сравнивать скорость стрельбы всех оружий в игре. Видно, что скорость стрельбы в 2 случаях скорость стрельбы повторяется целых 3 раза

```python
plt.figure(figsize=(5,5))
plt.hist(valorant["Wall Penetration"], bins = 6, edgecolor = "black", color = "pink")
plt.title("Распределение простреливаемости")
plt.xlabel("Уровень простреливаемости")
plt.ylabel("Количество значения")
plt.show()
```

Создаём ещё одну гистограмму в которой будем сравнивать уровни простреливаемости у всех оружий в игре. Видно, что чаще всего это значение 2, оно есть у 8 оружий в игре

```python
plt.figure(figsize=(6,6))
plt.hist(valorant["Magazine Capacity"], bins=100, edgecolor = "black", color = "orange")
plt.title("Распределение объёма магазина")
plt.xlabel("Объём магазина")
plt.ylabel("Количество значения")
plt.show()
```

Создаём ещё одну гистограмму в которой будем сравнивать объём магазина у всех оружий в игре. Видно, что значения от 0 до 40 попадаются чаще всего
